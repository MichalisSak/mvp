```python
import pandas as pd
from unidecode import unidecode

year = 2022

mvp_df = pd.read_html(f"https://www.basketball-reference.com/awards/awards_{year}.html")[0]
mvp_df.columns = mvp_df.columns.droplevel()
mvp_df.rename(columns={'Tm': 'Team'}, inplace=True)
mvp_df.rename(columns={'Share': 'VPS'}, inplace=True)
mvp_df.rename(columns={'WS': 'T_WS'}, inplace=True)
mvp_df = mvp_df.drop(['Age','First','Pts Won','Pts Max'], axis = 1)
mvp_df = mvp_df.head(10)
mvp_df
```

```python
mvp_df[['T_PTS', 'T_AST', 'T_TRB', 'T_STL', 'T_BLK', 'T_TOV']] = None
mvp_df
```

```python
playerNames = mvp_df['Player'].tolist()
i = 0
for playerName in playerNames:
    first_name, last_name = unidecode(playerName).lower().split()
    player_url = f"https://www.basketball-reference.com/players/{last_name[0]}/{last_name[:5]}{first_name[:2]}01.html"
    player_url2 = f"https://www.basketball-reference.com/players/{last_name[0]}/{last_name[:5]}{first_name[:2]}02.html"
    
    try:
        players_total_stats_df = pd.read_html(player_url)[4]
        players_total_stats_df = players_total_stats_df[['Season', 'PTS', 'TRB', 'AST', 'STL', 'BLK', 'TOV']]
        stats_of_season = players_total_stats_df[players_total_stats_df['Season'] == f'{year-1}-{year-2000}'].iloc[0]
        
    except:
        players_total_stats_df = pd.read_html(player_url2)[4]
        players_total_stats_df = players_total_stats_df[['Season', 'PTS', 'TRB', 'AST', 'STL', 'BLK', 'TOV']]
        stats_of_season = players_total_stats_df[players_total_stats_df['Season'] == f'{year-1}-{year-2000}'].iloc[0]
        
    players_total_stats_list = stats_of_season.to_list()
    players_total_stats_list = players_total_stats_list[1:]
    
    mvp_df.loc[i, ['T_PTS', 'T_TRB', 'T_AST', 'T_STL', 'T_BLK', 'T_TOV']] = [int(stat) for stat in players_total_stats_list]
    i += 1
    
mvp_df
```

```python
mvp_df = mvp_df[['Player','T_PTS', 'T_TRB', 'T_AST', 'T_STL', 'T_BLK', 'T_TOV', 'T_WS']]
mvp_df
```

```python
mvp_df.to_csv(f'mvp{year}totals.csv', index=False)
```
