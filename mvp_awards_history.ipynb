```python
import pandas as pd
from unidecode import unidecode

year = 2022

mvp_df = pd.read_html(f"https://www.basketball-reference.com/awards/awards_{year}.html")[0]
mvp_df.columns = mvp_df.columns.droplevel()
mvp_df.rename(columns={'Tm': 'Team'}, inplace=True)
mvp_df.rename(columns={'Share': 'VPS'}, inplace=True)
mvp_df = mvp_df.drop(['Age','First','Pts Won','Pts Max'], axis = 1)
mvp_df = mvp_df.head(10)
```

```python
nba_teams = {
    "ATL": "Atlanta Hawks",
    "BOS": "Boston Celtics",
    "BRK": "Brooklyn Nets",
    "CHO": "Charlotte Hornets",
    "CHA": "Charlotte Bobcats",
    "CHI": "Chicago Bulls",
    "CLE": "Cleveland Cavaliers",
    "DAL": "Dallas Mavericks",
    "DEN": "Denver Nuggets",
    "DET": "Detroit Pistons",
    "GSW": "Golden State Warriors",
    "HOU": "Houston Rockets",
    "IND": "Indiana Pacers",
    "LAC": "Los Angeles Clippers",
    "LAL": "Los Angeles Lakers",
    "MEM": "Memphis Grizzlies",
    "MIA": "Miami Heat",
    "MIL": "Milwaukee Bucks",
    "MIN": "Minnesota Timberwolves",
    "NOP": "New Orleans Pelicans",
    "NOH": "New Orleans Hornets",
    "NYK": "New York Knicks",
    "OKC": "Oklahoma City Thunder",
    "ORL": "Orlando Magic",
    "PHI": "Philadelphia 76ers",
    "PHO": "Phoenix Suns",
    "POR": "Portland Trail Blazers",
    "SAC": "Sacramento Kings",
    "SAS": "San Antonio Spurs",
    "TOR": "Toronto Raptors",
    "UTA": "Utah Jazz",
    "WAS": "Washington Wizards"
}
```

```python
tdfi = 10
if year <= 2015:
    tdfi =8
teams_df = pd.read_html(f"https://www.basketball-reference.com/leagues/NBA_{year}.html")[tdfi]
teams_df.columns = teams_df.columns.droplevel()
teams_df["Team"] = teams_df["Team"].str.replace("*", "")
teams_df = teams_df.sort_values(by=['W'], ascending=False, ignore_index=True)
teams_df['Team'] = teams_df['Team'].replace(nba_teams.values(), nba_teams.keys())
teams_df = teams_df[['Team', 'W']]
teams_df = teams_df.drop(30)
teams_df ['TWR'] = round(teams_df['W'] / teams_df.iloc[0, teams_df.columns.get_loc('W')], 2)
teams_df
```

```python
mvp_df = mvp_df.merge(teams_df, on='Team', how='left')
mvp_df = mvp_df.drop(['W'], axis = 1)
mvp_df[['PER', 'TS%', 'BPM', 'VORP', 'On_Off', 'TOV']] = None
mvp_df
```

```python
playerNames = mvp_df['Player'].tolist()
i = 0
for playerName in playerNames:
    first_name, last_name = unidecode(playerName).lower().split()
    player_url = f"https://www.basketball-reference.com/players/{last_name[0]}/{last_name[:5]}{first_name[:2]}01.html"
    player_url2 = f"https://www.basketball-reference.com/players/{last_name[0]}/{last_name[:5]}{first_name[:2]}02.html"
    
    try:
        player_on_off_url = player_url[:-5] + f"/on-off/{year}"
        on_off_df = pd.read_html(player_on_off_url)[0]
        on_off_value = on_off_df.iloc[-1, -1]
        
        players_advanced_stats_df = pd.read_html(player_url)[6]
        players_advanced_stats_df = players_advanced_stats_df[['Season', 'PER', 'TS%', 'BPM', 'VORP']]
        stats_of_season = players_advanced_stats_df[players_advanced_stats_df['Season'] == f'{year-1}-{year-2000}'].iloc[0]
        
    except:
        player_on_off_url = player_url2[:-5] + f"/on-off/{year}"
        on_off_df = pd.read_html(player_on_off_url)[0]
        on_off_value = on_off_df.iloc[-1, -1]

        players_advanced_stats_df = pd.read_html(player_url2)[6]
        players_advanced_stats_df = players_advanced_stats_df[['Season', 'PER', 'TS%', 'BPM', 'VORP']]
        stats_of_season = players_advanced_stats_df[players_advanced_stats_df['Season'] == f'{year-1}-{year-2000}'].iloc[0]
        
    players_advanced_stats_list = stats_of_season.to_list()
    players_advanced_stats_list = players_advanced_stats_list[1:]
    players_advanced_stats_list.append(on_off_value)
    mvp_df.loc[i, ['PER', 'TS%', 'BPM', 'VORP', 'On_Off']] = players_advanced_stats_list
    i += 1
mvp_df
```

```python
i = 0
for playerName in playerNames:
    first_name, last_name = unidecode(playerName).lower().split()
    player_url = f"https://www.basketball-reference.com/players/{last_name[0]}/{last_name[:5]}{first_name[:2]}01.html"
    player_url2 = f"https://www.basketball-reference.com/players/{last_name[0]}/{last_name[:5]}{first_name[:2]}02.html"
    
    try:
        tovdf = pd.read_html(player_url)[1]
        tov = tovdf.loc[tovdf['Season'] == f'{year-1}-{year-2000}', 'TOV'].iloc[0]
        
    except:
        tovdf = pd.read_html(player_url2)[1]
        tov = tovdf.loc[tovdf['Season'] == f'{year-1}-{year-2000}', 'TOV'].iloc[0]
        
    mvp_df.loc[i, ['TOV']] = float(tov)
    i += 1
mvp_df
```

```python
print(mvp_df['TOV'].dtype)
mvp_df['TOV'] = mvp_df['TOV'].astype('float64')
print(mvp_df['TOV'].dtype)
```

```python
per36 = ['PTS', 'TRB','AST', 'BLK', 'STL', 'TOV']
for stat in per36:
    mvp_df = mvp_df.rename(columns={f'{stat}': f'{stat}/36'})
    mvp_df[f'{stat}/36'] = round((36 * mvp_df[f'{stat}/36']) / mvp_df['MP'], 1)
mvp_df
```

```python
mvp_df = mvp_df.rename(columns={'WS': 'WS/36'})
mvp_df['WS/36'] = (36 * mvp_df['WS/36']) / (mvp_df['G'] * mvp_df['MP'])
mvp_df['WS/36'] = mvp_df['WS/36'].apply(lambda x: round(x, 3))
mvp_df
```

```python
mvp_df = mvp_df[['Player', 'G', 'PTS/36', 'TRB/36', 'AST/36', 'STL/36', 'BLK/36', 'TOV/36', 'TS%', 'PER', 'WS/36', 'BPM', 'VORP', 'On_Off', 'TWR', 'VPS']]
mvp_df
```

```python
# Derrick Rose in 2021 was traded from DET to NYK. He was the only player ending up on the top 10 for years 2013-2022. So his team stats are for NYK.
if year==2021:
    mvp_df.loc[mvp_df['Player'] == 'Derrick Rose', 'TWR'] = 0.79
mvp_df
```

```python
mvp_df.to_csv(f'mvp{year}.csv', index=False)
```
